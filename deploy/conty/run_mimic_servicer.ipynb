```python
import os
import re
import sys
base_deploy_dir = re.search(r'(.*/deploy/)', os.path.abspath("data_example.ipynb")).group(1)
sys.path.append(base_deploy_dir)

import grpc
from concurrent import futures
from conty.mimic_servicer import *
```

```python
# Start MoveMimicServicer
server = grpc.server(futures.ThreadPoolExecutor(max_workers=10),
                     options=
                     [('grpc.max_send_message_length', 10 * 1024 * 1024),
                      ('grpc.max_receive_message_length', 10 * 1024 * 1024)]
                     )
servicer = MoveMimicPC_servicer()
mimic_grpc.add_MoveMimicServicer_to_server(servicer=servicer, server=server)

server.add_insecure_port('[::]:{}'.format(PC_DEVICE_PORT))
server.start()
```
