```python
import cv2
import os
import numpy as np
import h5py
import matplotlib.pyplot as plt
import re

match = re.search(r'(.*/deploy/)', os.path.abspath("data_example.ipynb"))
base_train_dir = os.path.join(match.group(1), "../train")
```

```python
dataset_path = os.path.join(base_train_dir, "data/broom", "0.h5")

with h5py.File(dataset_path, "r") as root:
    print(root.keys())
    time_idx = 0
    image = root["images.rgb.left"][time_idx]

print(image.shape)
plt.imshow(image)
```

```python
crop_area = [[100, 400], [20, 600]]
cropped_image = image[crop_area[0][0]:crop_area[0][1], crop_area[1][0]:crop_area[1][1]]
print(cropped_image.shape)
plt.imshow(cropped_image)
```
